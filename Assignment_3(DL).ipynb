```python
# Step 2: Separate Inputs and Output
X = data[["experience", "age"]].values
y = data["income"].values
# X shape: (N, 2)
# y shape: (N,)
```

```python
# Step 3: Initialize the Model Parameters
n_features = X.shape[1]
w = np.zeros(n_features)
b = 0.0
# Why do we need one weight per feature?
# Each feature contributes independently to the output. A separate weight lets the model learn how important each feature is.

# Why is bias separate?
# Bias shifts the prediction up or down independent of input features. Without bias, the model would be forced to pass through the origin.

# Would initializing with large values be risky?
# Yes. Large initial weights produce large predictions and errors, causing unstable gradients and possibly preventing convergence.
```

Initializes:

and one bias term

Why it is needed

Learning starts from an initial guess.
Gradient descent will improve these values step by step.

```python
# Step 4: Define the Forward Pass (Prediction)
def predict(X, w, b):
    y_hat = X.dot(w) + b
    return y_hat
# Why is there no activation function?
# This is a regression problem where the output is continuous. Activation functions are mainly needed for classification, not numeric prediction.

# What kind of values can y_hat take?
# y_hat can take any real-valued number (positive, negative, or zero).

# How is this different from logistic regression?
# Linear regression predicts continuous values with no activation function, while logistic regression uses a sigmoid activation and outputs probabilities between 0 and 1.
```

```python
# Step 5: Define the Loss Function (MSE)
def mean_squared_error(y, y_hat):
    loss = ((y_hat - y) ** 2).mean()
    return loss
# Why square the error?
# Squaring makes all errors positive and penalizes large errors more strongly.

# What happens if one prediction is very wrong?
# Its squared error becomes very large, increasing the loss significantly and forcing the model to correct that mistake.

# Why not just take absolute error?
# Absolute error is not differentiable at zero, making gradient-based optimization harder compared to squared error.
```

Calculates how wrong predictions are on average.

Why it is needed:
The model needs a numeric measure of “how bad” its predictions are.

```python
# Step 6: Compute Gradients (Core Learning Step)
def compute_gradients(X, y, y_hat):
    N = len(y)

    dw = (2 / N) * X.T.dot(y_hat - y)
    db = (2 / N) * (y_hat - y).sum()

    return dw, db
# Why does X appear in dw but not in db?
# Weights multiply the input features, so their gradients depend on X. Bias is independent of inputs, so X does not appear in db.

# Why does the error term appear everywhere?
# Gradients are proportional to the prediction error. If there is no error, there is nothing for the model to correct.

# What happens if error is zero?
# Loss becomes zero and gradients become zero, so weights and bias stop updating (learning stops).
```

Computes how much each parameter contributes to error.

Why it is needed
(Without gradients):

We don’t know how to update weights

Learning cannot happen

```python
# Step 7: Update Parameters (Gradient Descent)
def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b
```

Moves weights and bias in the direction that reduces loss.

Why it is needed:
This is the learning rule of the perceptron.

```python
# Step 8: Training Loop
lr = 0.0001
epochs = 1000

for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")
# Does loss decrease over time?
# Yes, if learning rate and gradients are correct, loss should decrease as the model learns better weights.

# What happens if loss increases?
# It usually means the learning rate is too high or gradients are incorrect, causing unstable updates.

# How do learning rate and epochs interact?
# Learning rate controls step size, epochs control number of updates. A small learning rate needs more epochs, while a large one needs fewer but risks divergence.
```

Repeats learning steps multiple times:

1. Predict

2. Measure error

3. Compute gradients

4. Update parameters

Why it is needed:
Learning is gradual. One update is never enough.

```python
# Step 9: Final Evaluation
print("Final weights:", w)
print("Final bias:", b)
new_candidate = np.array([4.5, 68])
predicted_salary = new_candidate.dot(w) + b
print("Predicted Salary:", predicted_salary)

# Is the prediction reasonable?
# Yes, the prediction lies within the range of training data and follows the learned trend between inputs and salary.

# Does it interpolate smoothly?
# Yes, the linear model produces smooth, continuous predictions between known data points.

# Why is this better than threshold rules?
# It learns relationships from data, uses all features together, and avoids rigid, manually defined thresholds.
```
