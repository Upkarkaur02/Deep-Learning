```python
# Task 1.1
import numpy as np
X = np.linspace(0, 10, 50).reshape(-1, 1)

# X represents the single input feature.
# The rangel 0-10 allows us to see early rapid growth and later slow saturation behavior.
# This helps demonstrate why a linear model fails.
```

Why it is needed: We want data that:
1. grows fast at first
2. slows later
A straight line cannot bend, so this rangle helps us see failure of linear models

```python
# Task 1.2 Target generation
noise = np.random.normal(0, 0.2, size=(50,1))
y = np.log(X + 1) + noise

# Log func. grows fast initially and slows down later.
# A straight line has only one sloppe so it cannot match both regions well
```

Why it is needed:

Because:  y= log(x+1)

This curve: 1. steep at beginning

2. Flattens later

```python
# Decide Model shape:
# We choose 3 hidden units

# Why more than 1?
# One hidden unit behaves like one piece of curve.
# Multiple units allow multiple "bends".

# Why not too many?
# Too many increases complexity, instability and overfitting
```

```python
# Initialize Model Parameters
W1 = np.random.uniform(-1, 1, size=(1,3))
b1 = np.zeros((1,3))
W2 = np.random.uniform(-1, 1, size=(3,1))
b2 = np.zeros((1,1))
```

```python
# Activation Function
def Activation(z):
  return np.maximum(0, z)

# keeps positive values and Supresses negative values
```

```python
# Activation Slope
def activation_slope(z):
  return(z>0).astype(float)

# slope = 1 if z>0
# slope = 0 if z<0

# Without slope the model has no direction signal telling it how to change parameters
```

```python
# Forward Pass
z1 = X @ W1 + b1
h = Activation(z1)
y_hat = h @ W2 + b2
```

```python
# Define Error and Loss
error = y_hat - y

loss = np.mean(error ** 2)

# Large error become much larger and model strongly correct big mistakes
```

```python
# Backward Flow
dL_dy = 2 * error/len(X)
```

```python
# Gradient of Output Layer
dL_dW2 = X.T @dL_dydL_db2 = np.sum(dL_dy, axis=0, keepdims=True)
```

Why depends on hidde values?

Because output weight multiplies hidden values.

If hidden unit is strong -> it affects output strongly

```python
# Flow into hidden layer
dL_dh = dL_dy @ W2.T
```

```python
# Apply Activation Slope
dL_dz1 = dL_dh * activation_slope(z1)


# How activation controls error flow
# Relu blocks negative signals
# if activation is OFF -> no gradient passes so, activation controls error
```

```python
# Gradient of First Layer
dL_dW1 = X.T @ dL_dz1
dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
```

```python
Learning_rate = 0.01
W1 -= Learning_rate * dL_dW1
b1 -= Learning_rate * dL_db1
W2 -= Learning_rate * dL_dW2
b2 -= Learning_rate * dL_db2

# We move opposite direction of gradient
```

```python
# Training Loop
Learning_rate = 0.01

for epoch in range(1000):

  # Forward
  z1 = X @ W1 + b1
  h = Activation(z1)
  y_hat = h @ W2 + b2

  # Loss
  error = y_hat - y
  loss = np.mean(error ** 2)

  # Backward
  dL_dy = 2 * error/len(X)
  dL_dW2 = h.T @dL_dy
  dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

  dL_dh = dL_dy @ W2.T
  dL_dz1 = dL_dh * activation_slope(z1)

  dL_dW1 = X.T @ dL_dz1
  dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

  # Update
  W1 -= Learning_rate * dL_dW1
  b1 -= Learning_rate * dL_db1
  W2 -= Learning_rate * dL_dW2
  b2 -= Learning_rate * dL_db2

  if epoch % 100 == 0:
    print(f"Epoch {epoch}, Loss: {loss}")

```

Why 1000 epochs?

-> Because Small dataset and Small learning rate Needs time to converge

Why Learning rate is 0.01?

-> Too large is unstable and too small is very slow 0.01 is moderate value
